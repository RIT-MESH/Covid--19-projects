```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

### 1. Import pandas and read in the india_statewise.json file into a dataframe called covid.

```python
df=pd.read_json('india_statewise.json')
```

```python
df
```

### 2. Normalize the dataframe with ‘data’ as column and ‘statewise’ as row and rename it as df_india

```python
df_india = pd.io.json.json_normalize(df['data']['statewise'])
df_india
```

### 3. Now make ‘state’ column as index

```python
df_india = df_india.set_index('state')
df_india.head()
```

### 4. Now Identify a row which is not a valid state and drop it

```python
df_india = df_india.drop([ 'State Unassigned'], axis=0)
df_india
```

### 5. Calculate the Mortality rate (deaths/confirmed) for each state and identify top 5 sates with highest mortality rate.

```python
df_india['Mortality Rate'] =(df_india['deaths']/df_india['confirmed'])*100
df_india
```

```python
df_india.nlargest(5, ['Mortality Rate']) #identify top 5 sates with highest mortality rate
```

### Now plot this data using folium map. The expected output should be is shown below:

```python
india = folium.Map(location=[23,80],tiles = 'cartodbpositron',zoom_start=4,max_zoom=8,min_zoom=3)

for i in range(0,len(df_india[df_india['confirmed']>0].index)):# range from 0 to length of index of df_india that to only if
                                                                #confirmed cases is greater than 0
    folium.Circle(
        location=[df_india.iloc[i]['Lat'],df_india.iloc[i]['Long']],#location of index in term of Latitude and Longitude
        tooltip = "<h5 style='text-align:Centre;font-weight:bold'>"+df_india.iloc[i].name+"</h5>"+#For sate name or index name style and font
        "<li>confirmed:"+str(df_india.iloc[i]['confirmed'])+"</li>"+# This we want to put in highlighted circle
        "<li>recovered:"+str(df_india.iloc[i]['recovered'])+"</li>"+# This we want to put in highlighted circle
        "<li>active:"+str(df_india.iloc[i]['active'])+"</li>"+# This we want to put in highlighted circle
        "<li>deaths:"+str(df_india.iloc[i]['deaths'])+"</li>"+# This we want to put in highlighted circle
        "<li>Mortality Rate:"+str(df_india.iloc[i]['Mortality Rate'])+"</li>",# This we want to put in highlighted cirle
        radius = (int(df_india.iloc[i]['confirmed'])), # radiu equivalent to confirmed cases
        color='#FF4500',#html colour codes oranged
        fill_color="#FF4500",
        fill=True).add_to(india)
```
