```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import folium
```

```python
covid19 = pd.read_csv('Covid19_India.csv')
```

```python
df_india = covid19.set_index('State')
df_india.head()
```

```python
df_india = df_india[['Confirmed','Recovered','Deaths','Active']]
df_india
```

```python
df_india['Lat'] = ""
df_india['Long'] = ""
for index in df_india.index:
    df_india.loc[df_india.index==index,'Lat']=locations[index][0]
    df_india.loc[df_india.index==index,'Long']=locations[index][1]
```

```python
df_india
```

```python
df_india['Mortality Rate'] = np.round(np.nan_to_num((df_india['Deaths']/df_india['Confirmed'])*100,2))
```

```python
df_india.head()
```

```python
india = folium.Map(location=[23,80],tiles = 'cartodbpositron',zoom_start=4,max_zoom=7,min_zoom=4,height=500,width='80%')

for i in range(0,len(df_india[df_india['Confirmed']>0].index)):
    folium.Circle(
        location=[df_india.iloc[i]['Lat'],df_india.iloc[i]['Long']],
        tooltip = "<h5 style='text-align:Centre;font-weight:bold'>"+df_india.iloc[i].name+"</h5>"+
        "<li>Confirmed:"+str(df_india.iloc[i]['Confirmed'])+"</li>"+
        "<li>Recovered:"+str(df_india.iloc[i]['Recovered'])+"</li>"+
        "<li>Active:"+str(df_india.iloc[i]['Active'])+"</li>"+
        "<li>Deaths:"+str(df_india.iloc[i]['Deaths'])+"</li>"+
        "<li>Mortality Rate:"+str(df_india.iloc[i]['Mortality Rate'])+"</li>",
        
        radius = (int(df_india.iloc[i]['Confirmed']+1)*2),
        color='#ff6600',
        fill_color="#ff6600",
        fill=True).add_to(india)
    
india
```
